```python
import os
import pandas as pd
import json
```

```python
data_home = '../data/'
data_name = 'the-movies-dataset'
data_path = os.path.join(data_home, data_name)
file_names = ['credits.csv',         #0
              'links.csv',           #1
              'movies_metadata.csv', #2
              'ratings_small.csv',   #3
              'keywords.csv',        #4
              'links_small.csv',     #5
              'ratings.csv'          #6
             ]
```

```python
dfs = [pd.read_csv(os.path.join(data_path, fn)) for fn in file_names]
```

```python
my_df = dfs[2][['id', 'release_date']]
my_df['release_date'] = my_df['release_date'].fillna('0000-00-00')
my_df['year'] = my_df['release_date'].apply(lambda x: int(x[:4]))
my_df = my_df[my_df['id'].apply(lambda x: '-' not in x)]
my_df['id'] = my_df['id'].apply(int)
my_df = my_df[['id', 'year']]
my_df.head(5)
```

```python
mcc_df = dfs[0].copy(deep=True)
mcc_df['cast'] = mcc_df['cast'].apply(lambda x: eval(x))
mcc_df['crew'] = mcc_df['crew'].apply(lambda x: eval(x))
mcc_df['cast_ids'] = mcc_df['cast'].apply(lambda x: [i['id'] for i in x])
mcc_df['crew_ids'] = mcc_df['crew'].apply(lambda x: [i['id'] for i in x])
mcc_df = mcc_df[['id', 'cast_ids', 'crew_ids']]
mcast_df = mcc_df[['id', 'cast_ids']]
mcast_df = mcast_df[mcast_df['cast_ids'].apply(lambda x: len(x) > 0)]
mcrew_df = mcc_df[['id', 'crew_ids']]
mcrew_df = mcrew_df[mcrew_df['crew_ids'].apply(lambda x: len(x) > 0)]
```

```python
mk_df = dfs[4].copy(deep=True)
mk_df['keywords'] = mk_df['keywords'].apply(lambda x: eval(x)).apply(lambda x: [int(i['id']) for i in x])
mk_df = mk_df[mk_df['keywords'].apply(lambda x: len(x) > 0)]
mk_df.head(5)
```
